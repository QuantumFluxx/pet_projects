```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.simplefilter('ignore')

%matplotlib inline
```

```python
movies = pd.read_csv('C:/Users/Александр/Projects/projects/Pet-projects/Top_1000_movies/movies_IMDB.csv')
movies.head()
```

```python
# Мы видим первую колонку с названием Unnamed: 0. Она не несет никакой полезной информации
movies.drop(columns=['Unnamed: 0'] , inplace=True)
movies.head()
```

```python
movies.describe().T
```

# Проведем разведочный анализ данных

```python
# Здесь мы можем увидеть, что жанр Драма преобладает над остальными
movies['genre'].value_counts().nlargest(10).plot(kind='pie' ,title= 'Фильмов на жанр в %', figsize=(10,10), autopct='%1.1f%%',fontsize=15)
sns.despine()
```

```python
# мы можем видеть, что больше всего фильмов было в 2004 году
movies['year'].value_counts().nlargest(40).plot(kind='bar', figsize=(13, 9))
sns.despine()
```

```python
# также мы можем получить лучшие 50 фильмов на основе рейтинга imdb

top_movies = movies.sort_values('imdb',ascending=False).head(50)
fig,ax = plt.subplots(figsize=(30, 7))
ax = sns.barplot(x='movie_name', y='imdb', data=top_movies,ci=None)
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
fig.suptitle('Топ фильмы за все время', fontsize=16)
ax.set_xlabel('Название фильма',fontsize=20)
ax.set_ylabel('Рейтинг IMDB',fontsize=20)
ax.tick_params(axis='x', labelsize=15)
ax.tick_params(axis='y', labelsize=15)
plt.show();
```

### Лучшие фильмы по рейтингу IMDB только за 2018 год

```python
all_top_movies= movies.sort_values('imdb',ascending=False)
top_2018 = all_top_movies[all_top_movies['year'] == 2018]
fig,ax = plt.subplots(figsize=(30, 7))
ax = sns.barplot(x='movie_name', y='imdb', data=top_2018,ci=None)
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
fig.suptitle('Tоп фильмы 2018 года', fontsize=18)
ax.set_xlabel('Название фильма',fontsize=20)
ax.set_ylabel('Рейтинг IMDB',fontsize=20)
ax.tick_params(axis='x', labelsize=15)
ax.tick_params(axis='y', labelsize=15)
plt.show();
```
